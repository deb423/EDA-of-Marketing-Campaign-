# Merging the marital status columns into 1 column by replacing 1 with a number and 0 with a blank and then joinging the columns

```python
#First, Change Column data type to string and then replace 1 with different number & 0 with blank.
df_iffoods['marital_Married']=df_iffoods['marital_Married'].astype(str).replace(['1','0'],['5',''])
df_iffoods['marital_Single']=df_iffoods['marital_Single'].astype(str).replace(['1','0'],['4',''])
df_iffoods['marital_Together']=df_iffoods['marital_Together'].astype(str).replace(['1','0'],['3',''])
df_iffoods['marital_Widow']=df_iffoods['marital_Widow'].astype(str).replace(['1','0'],['2',''])
df_iffoods['marital_Divorced']=df_iffoods['marital_Divorced'].astype(str).replace(['1','0'],['1',''])
#Now all columns contain different numbers for different marital status, lets join them in one column.
df_iffoods['marital_status']=df_iffoods['marital_Widow']+df_iffoods['marital_Together']+df_iffoods['marital_Single']+df_iffoods['marital_Married']+df_iffoods['marital_Divorced']
#Next, we map numbers into different categorical values.
df_iffoods['marital_status']=df_iffoods['marital_status'].map({'1':'Divorced', '2':'Widow','3':'Together','4':'Single','5':'Married'})
```

```python
df_iffoods['marital_status'].unique()
```

```python
df_iffoods.shape
```

```python
# dropping the marital status columns as we have already joined them together under marital_status 
df_mktdata =df_iffoods.drop(['marital_Widow','marital_Together','marital_Single','marital_Married','marital_Divorced'],axis =1)
df_mktdata.columns
```

# Joining the education column after replacing the values 

```python
df_mktdata['education_2n Cycle']=df_mktdata['education_2n Cycle'].astype(str).replace({'0':''})
df_mktdata['education_Basic']=df_mktdata['education_Basic'].astype(str).replace({'1':'2','0':''})
df_mktdata['education_Graduation']=df_mktdata['education_Graduation'].astype(str).replace({'1':'3','0':''})
df_mktdata['education_Master']=df_mktdata['education_Master'].astype(str).replace({'1':'4','0':''})
df_mktdata['education_PhD']=df_mktdata['education_PhD'].astype(str).replace({'1':'5','0':''})
df_mktdata['education_level']=df_mktdata['education_2n Cycle']+df_mktdata['education_Basic']+df_mktdata['education_Graduation']+df_mktdata['education_Master']+df_mktdata['education_PhD']
df_mktdata['education_level']=df_mktdata['education_level'].map({'1':'2n Cycle','2':'Basic','3':'Graduation','4':'Master','5':'PhD'})
```

```python
df_mktdata['education_level'].unique()
```

```python
# Joining the KidHome and TeenHome columns to find the total children in a house 
df_mktdata['kids'] = df_mktdata['Kidhome'] + df_mktdata['Teenhome']
```

**Analysisng campaign with categorical values**

```python
def Cat_Campaign_Relation(df, column_name):
    e1=(df_mktdatafinal.groupby([column_name]).sum()['AcceptedCmp1']/df_mktdatafinal.groupby([column_name]).count()['AcceptedCmp1'])
    e2=(df_mktdatafinal.groupby([column_name]).sum()['AcceptedCmp2']/df_mktdatafinal.groupby([column_name]).count()['AcceptedCmp2'])
    e3=(df_mktdatafinal.groupby([column_name]).sum()['AcceptedCmp3']/df_mktdatafinal.groupby([column_name]).count()['AcceptedCmp3'])
    e4=(df_mktdatafinal.groupby([column_name]).sum()['AcceptedCmp4']/df_mktdatafinal.groupby([column_name]).count()['AcceptedCmp4'])
    e5=(df_mktdatafinal.groupby([column_name]).sum()['AcceptedCmp5']/df_mktdatafinal.groupby([column_name]).count()['AcceptedCmp5'])
    df_new=pd.concat([e1,e2,e3,e4,e5],axis=1).T
    plt.figure(figsize=(8,6))
    sns.lineplot(data=df_new, markers=True, linewidth=2)
    plt.ylabel('Percentage Acceptance')
    plt.show()
```
